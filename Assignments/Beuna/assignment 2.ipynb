# 5.Find the outliers and replace the outliers

```python
sns.boxplot(x=df["EstimatedSalary"])
```

```python
sns.boxplot(x=df["CreditScore"])
```

```python
q1=df.CreditScore.quantile(0.25)
q2=df.CreditScore.quantile(0.75)
IQR = q2-q1
upper_limit = q2+1.5*IQR
lower_limit = q2-1.5*IQR
print(upper_limit)
print(lower_limit)
df["CreditScore"].head()
```

```python
df.median()
```

```python
df["CreditScore"]=np.where(df["CreditScore"]<lower_limit,6.52,df["CreditScore"])
```

```python
df["CreditScore"]=np.where(df["CreditScore"]<lower_limit,6.52,df["CreditScore"])
sns.violinplot(x=df["CreditScore"])
```

# 6.Check for categorial columns and perform encoding

```python
df.head()
```

```python
le=LabelEncoder()
df.Geography=le.fit_transform(df.Geography)
df.Gender=le.fit_transform(df.Gender)
df.Surname=le.fit_transform(df.Surname)
```

# 7.Split the data

```python
y=df["EstimatedSalary"]
x=df.drop("EstimatedSalary",axis=1)
```

# 8.Scale the independent variable

```python
scaled_x=pd.DataFrame(scale(x),columns=x.columns)
scaled_x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(scaled_x,y,test_size=0.3,random_state=0)
print(xtrain.shape,xtest.shape,ytrain.shape,ytest.shape)
```
